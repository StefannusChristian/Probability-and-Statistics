#Masalah Manajemen CSO

Suatu bank mengelola 85 karyawan CSO (Customer Service Officer) untuk melayani kebutuhan nasabah. Sistem informasi CSO mencatat dua hal berikut:

Banyaknya customer yang pernah dilayani masing-masing CSO
Di akhir sesi layanan, customer akan menekan tombol 'like' atau 'dislike' untuk menilai performa setiap CSO. Sistem informasi mencatat persentase customer yang menekan tombol 'dislike' di akhir sesi, menandakan persentase ketidakpuasan pelanggan atas pelayanan seorang CSO.

Hitunglah peluang kejadian berikut:

1. Berdasarkan data, berapakah peluang seorang nasabah secara umum tidak puas pada pelayanan CSO?
<br>
2. Ketika ditemukan seorang nasabah tidak puas terhadap pelayanan CS (namun tidak diketahui siapa tepatnya CSO yang melayani nasabah itu saat itu), siapakah CSO yang:
<br>
    a. Berpeluang paling tinggi untuk memberikan pelayanan pada nasabah tersebut waktu itu?
<br>
    b. Berpeluang paling rendah untuk memberikan pelayanan pada nasabah tersebut waktu itu?

## Import Data

```python
import pandas as pd
data = pd.read_csv("CSO_problem.csv")
data
```

## Nomor 1
Berdasarkan data, berapakah peluang seorang nasabah secara umum tidak puas pada pelayanan CSO?

```python
total_cust = sum(data['#served_cust'])
data['served_cust_per_total'] = [(cust/total_cust) for cust in data['#served_cust']]
data
```

```python
percent_unsatisfed = data['prcnt_unsatisfied']
percent_unsatisfed = [float(i[:-1])/100 for i in percent_unsatisfed]
served_cust = data['served_cust_per_total']
hasil = sum([percent_unsatisfed[i]*served_cust[i] for i in range(len(data))])
peluang_nomor_1 = hasil*100
print(f'Berdasarkan data, peluang seorang nasabah secara umum tidak puas pada pelayanan CSO adalah {round(peluang_nomor_1,3)}%')
```

## Nomor 2
Ketika ditemukan seorang nasabah tidak puas terhadap pelayanan CS (namun tidak diketahui siapa tepatnya CSO yang melayani nasabah itu saat itu), siapakah CSO yang:

```python
data['unsatisfied_x_#served_cust'] = [percent_unsatisfed[i]*served_cust[i] for i in range(len(data))]
data
```

```python
# peluang cso | tidak puas
# rumus: (peluang_cso * tidak_puas bersyarat CSO) / tidak_puas 
unsatisfied_x_ = data['unsatisfied_x_#served_cust']
data['peluang cso | tidak puas'] = [(served_cust[i]*unsatisfied_x_[i])/hasil for i in range(len(data))] 
data.sort_values(by=['peluang cso | tidak puas'],inplace=True,ascending=False)
data
```

### 2a dan b
a. Berpeluang paling tinggi untuk memberikan pelayanan pada nasabah tersebut waktu itu?
b. Berpeluang paling rendah untuk memberikan pelayanan pada nasabah tersebut waktu itu?

```python
cso_high_untuk_memberikan_pelayanan_pada_nasabah = data.head(1)['CSO'][48]
peluang_cso_high = data.head(1)['peluang cso | tidak puas'][48]

data_cso_low_untuk_memberikan_pelayanan_pada_nasabah = data[data['peluang cso | tidak puas']==0]
peluang_cso_low = data.tail(1)['peluang cso | tidak puas'][0]
cso_low_untuk_memberikan_pelayanan_pada_nasabah = ', '.join(list(data['CSO']))
peluang_cso_low = data.tail(1)['peluang cso | tidak puas'][0]

print(f'CSO yang berpeluang paling tinggi untuk memberikan pelayanan pada nasabah tersebut waktu itu adalah {cso_high_untuk_memberikan_pelayanan_pada_nasabah} dengan peluang {peluang_cso_high*100}%')
print(f'CSO yang berpeluang paling rendah untuk memberikan pelayanan pada nasabah tersebut waktu itu adalah:')
print(cso_low_untuk_memberikan_pelayanan_pada_nasabah)
print(f'Dengan peluang {peluang_cso_low}%')
```
